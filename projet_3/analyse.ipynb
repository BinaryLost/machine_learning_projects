```python
#import des fichiers et load des ressources
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

data = pd.read_csv('training_set.csv')
pd.options.display.max_columns = data.shape[1]
```

```python
data.sample(1)
```

```python
#création d'une colonne ancienneté à partir des variable DataYear et YearBuilt
data['Ancienneté'] = data['DataYear'] - data['YearBuilt']
data['Ancienneté'].unique()

bins = np.arange(0, data['Ancienneté'].max() + 10, 10)
plt.hist(data['Ancienneté'], bins=bins, edgecolor='black')
plt.xticks(bins)
plt.xlabel("Ancienneté")
plt.ylabel("Fréquence")
plt.title("Histogramme de l'ancienneté")
mean = data['Ancienneté'].mean()
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label='moyenne')
median = data['Ancienneté'].median()
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label='mediane')
plt.legend()
plt.show()

plt.boxplot(data['Ancienneté'], showmeans=True)
plt.show()
```
